## 2、从一个文件提取信息

### 2.1 获取Word中的第一个表格

- 获取文件夹中的文件名列表
- 获取文件夹中的第一个文件
- 读取第一个文件中的第一个表格，以dataframe形式进行查看

```python
import docx
import pandas as pd
import os

# 获取文件夹中的文件名列表
file_list = os.listdir(folder_path)

# 获取文件夹中的第一个文件
first_file = file_list[0]

# 读取第一个文件中的第一个表格
doc = docx.Document(os.path.join(folder_path, first_file))
table = doc.tables[0]

# 提取表格中的数据形成df，因为是不规则表格，所以行与列的索引都用整数
data = []
for i, row in enumerate(table.rows):
    row_data = []
    for j, cell in enumerate(row.cells):
        row_data.append(cell.text)
    data.append(row_data)

# 将数据转换为DataFrame
df = pd.DataFrame(data)

print(df)
```

## 3、批量操作文件夹中的文件

### 3.1 将一次性代码封装为一个函数，并循环处理文件夹中的所有文件

调整为一个函数与for循环，从文件夹中循环处理，并最终形成一个大的df

```python
import docx
import pandas as pd
import os

def extract_data_from_table(table):
    data = []
    for i, row in enumerate(table.rows):
        row_data = []
        for j, cell in enumerate(row.cells):
            row_data.append(cell.text)
        data.append(row_data)
    df = pd.DataFrame(data)
    
    extracted_data = {
        '姓名': df.iloc[1, 1],
        '性别': df.iloc[1, 3],
        '年龄': df.iloc[1, 5],
        '所在城市': df.iloc[2, 1],
        '工作单位': df.iloc[2, 3],
        '身高': df.iloc[3, 1],
        '体重（斤）': df.iloc[3, 3],
        '学历': df.iloc[3, 5],
        '性格特点': df.iloc[4, 1],
        '兴趣爱好': df.iloc[5, 1],
        '对方性别': df.iloc[7, 1],
        '年龄区间': df.iloc[7, 3],
        '备注': df.iloc[7, 5]
    }
    
    return pd.DataFrame([extracted_data])

# 获取文件夹中的文件名列表
file_list = os.listdir(folder_path)

# 初始化一个空的DataFrame
df_all = pd.DataFrame()

# 循环处理文件夹中的每个文件
for file_name in file_list:
    doc = docx.Document(os.path.join(folder_path, file_name))
    table = doc.tables[0]
    df_new = extract_data_from_table(table)
    df_all = pd.concat([df_all, df_new], ignore_index=True)

df_all
```

### 3.2 将提取的数据保存到Excel文件中

```python
# 保存到Excel文件
df_all.to_excel('报名表数据.xlsx', index=False)
```
